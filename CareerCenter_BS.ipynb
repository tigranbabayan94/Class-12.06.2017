```python
import requests
```

```python
from BeautifulSoup import * 
```

```python
url = "https://www.careercenter.am/ccidxann.php"
```

```python
response = requests.get(url)
page = response.text
soup = BeautifulSoup(page)
```

```python
tables = soup.findAll('table')
```

```python
type(tables)
```

```python
our_table = tables[0]
```

```python
rows = our_table.findAll('tr')
```

```python
rows
```

```python
data = []
for i in rows:
    columns = i.findAll("td")
    for j in columns:
         data.append(j.text)
```
